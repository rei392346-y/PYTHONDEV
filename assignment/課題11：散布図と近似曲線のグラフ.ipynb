```python
# 課題11：散布図と近似曲線のグラフ.ipynb

import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib # 日本語表示のためにインポート

# データの定義
x = np.array([62, 95, 83, 100, 72, 71, 55, 85, 67, 75])   # 英語の点数
y = np.array([62, 80, 85, 95, 58, 76, 52, 84, 67, 71])    # 数学の点数

# 散布図の作成
# 横軸をx (英語)、縦軸をy (数学) とし、点は青色で表示
plt.scatter(x, y, color='blue', label='score')

# 最小二乗法による近似直線の計算
# xとyから1次関数 (直線) の係数 (傾きと切片) を求める
# polyfit(xデータ, yデータ, 次数)
coefficients = np.polyfit(x, y, 1)
# 傾きと切片を取得
slope = coefficients[0]
intercept = coefficients[1]

# 近似直線のy値を計算
# xの範囲で直線のy値を生成
regression_line_y = slope * x + intercept

# 近似直線の描画
# 直線は赤色で表示
plt.plot(x, regression_line_y, color='red', label='lsm')

# グラフのラベル設定
plt.xlabel('english') # 横軸のラベルを 'english' に設定
plt.ylabel('math')    # 縦軸のラベルを 'math' に設定

# 凡例の表示
# 凡例を左上に表示
plt.legend(loc='upper left')

# グラフのタイトル
plt.title('散布図と近似直線を用いた英語と数学の点数比較')

# グラフの表示
plt.grid(True) # グリッド線を表示
plt.show()
```
